```python
import torch
import matplotlib.pyplot as plt
import math
```

# Problem 1:

```python
def genData(N):
    rawDat = torch.randn((N,101))
    rawOut = torch.ones((N))
    rawDat[:,0] = 1 
    for i in range(N):
        noise = ( torch.randn( 1 ) * 0.1 ).item()
        for j in range(51,61):
            rawDat[i,j] = rawDat[i,1]+0.5*rawDat[i,j-50] + noise
        for j in range(61,71):
            rawDat[i,j] = rawDat[i,j-60] - rawDat[i,j-50] + rawDat[i,40-j] + noise
        for j in range(71,81):
            rawDat[i,j] = rawDat[i,6*(j-70)]+3*rawDat[i,j-70]+noise
        for j in range(81,91):
            rawDat[i,j] = 5-rawDat[i,j-10]
        for j in range(91,101):
            rawDat[i,j] = 0.5*(rawDat[i,50+(j-90)*4]+rawDat[i,50+(j-90)*3]) + noise
        rawOut[i] = sum([rawDat[i,2*k]*(-0.88)**k for k in range(1,51)])
    return rawDat, rawOut
```

```python
print(genData(2))
```

## Problem 2:

```python
def pred(X,w):
    return torch.matmul(X,w)
```

```python
def lossFunc(X,w,yreal):
    return torch.mean(torch.square(yreal - pred(X,w)))
```

## Problem 3:

```python
def train(Xtrain,ytrain,Xtest,ytest,a,quiet):
    LossTrain = []
    LossTest = []
    w = torch.nn.Parameter( torch.randn((101)), requires_grad = True )
    loss = lossFunc(Xtrain,w,ytrain)
    lossT = lossFunc(Xtest,w,ytest).item()
    i = 0
    diff = 0
    while loss >0.001 and i < 1000000 and diff < 1:
        loss = lossFunc(Xtrain,w,ytrain)
  
        loss.backward()
        
        with torch.no_grad():
            w -= a * w.grad
        
        oldLossT = lossT
        lossT = lossFunc(Xtest,w,ytest).item()
        diff = lossT - oldLossT
        
        w.grad = None
        LossTrain.append(loss.item())
        LossTest.append(lossT)
        if i % 100 == 0 and not quiet:
            print(i+1,"Training Loss:",LossTrain[i],"Testing Loss",LossTest[i])
        i += 1
    if diff > 1:
        print("loss got too large with alpha:",a)
    return w,LossTrain,LossTest
    
```

The conditions to end training are threefold.  The first is to see whether the loss gets sufficently small.  The second is to end 
training once it goes on too long.  The third is to detect if the testing loss is increasing unaccetably

```python
def vis_loss(lTrain,lTest):
  plt.figure()
  plt.plot(lTrain)
  plt.title("Training Loss")
  plt.show()

  plt.figure()
  plt.plot(lTest)
  plt.title("Testing Loss")
  plt.show()
vis_loss(lTrain,lTest)
```

As you see in my program, bad $\alpha$ values are taken out early by observing that the testing loss should be roughly going 
down over time.  Additionally since the loss is quadratic then when the loss increases, it will potentially eliminate itself
in the first 10 terms.

When we have a convergent $\alpha$ then both the testing loss and training loss go down in what appears to be an asymptotic 
manor.  However typically the testing loss is an order of magnitude or two greater than the training loss.

We observe that $\alpha = 0.0001$ is a good enough $\alpha$ value, and for a random vector in our data set we have that our 
"error" is

## Problem 6:

```python
def ridgeLoss(X,w,y,l):
    ridgeTerm = 0
    for i in range(1,101):
        ridgeTerm += (w[i].item())**2 
    return lossFunc(Xtrain,w,ytrain) + l*ridgeTerm
```

```python
def ridgeTrain(Xtrain,ytrain,Xtest,ytest,a,l,quiet):
    LossTrain = []
    LossTest = []
    w = torch.nn.Parameter( torch.randn((101)), requires_grad = True )
    loss = ridgeLoss(Xtrain,w,ytrain,l)
    lossTestV = lossFunc(Xtest,w,ytest).item()
    i = 0
    diff = 0
    while i < 10000:
        oldLoss = loss
        loss = ridgeLoss(Xtrain,w,ytrain,l)
        diff = loss.item() / oldLoss.item() 
        loss.backward()
        
        with torch.no_grad():
            w -= a * w.grad
        w.grad = None
        LossTrain.append(loss.item())
        LossTest.append(lossTestV)
        lossTestV = lossFunc(Xtest,w,ytest).item()
        if i+1 % 100 == 0 and not quiet:
            print(i+1,"Training Loss:",LossTrain[i],"Testing Loss",LossTest[i])
        i += 1
    if diff > 10:
        print("loss got too large with alpha:",a)
    return w,LossTrain,LossTest
```
